```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# Load the data
df = pd.read_csv('Lottery_Mega_Millions_Winning_Numbers__Beginning_2002.csv')

df = df['Winning Numbers'].str.split(' ', expand=True)

# Rename the new columns to more meaningful names
df.columns = ['Num1', 'Num2', 'Num3', 'Num4', 'Num5']
# Display the first few rows of the dataframe
print(df.head())

# Preprocessing
# Assume the columns are 'Num1', 'Num2', 'Num3', 'Num4', 'Num5', 'Powerball'
# We will predict the maximum numbers for each position (Num1 to Num5 and Powerball)
df['MaxNum1'] = df['Num1'].max()
df['MaxNum2'] = df['Num2'].max()
df['MaxNum3'] = df['Num3'].max()
df['MaxNum4'] = df['Num4'].max()
df['MaxNum5'] = df['Num5'].max()
#df['MegaBall'] = df['MegaBall'].max()

# Feature and target selection
X = df[['Num1', 'Num2', 'Num3', 'Num4', 'Num5']]
y = df[['MaxNum1', 'MaxNum2', 'MaxNum3', 'MaxNum4', 'MaxNum5']]

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Create and train the model
model = DecisionTreeRegressor(random_state=0)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Display predictions
for i in range(len(X_test)):
    print(f"Input: {X_test.iloc[i].values}, Predicted Max Numbers: {y_pred[i]}")
```

```python
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
import numpy as np

# Calculate absolute errors
absolute_errors = np.abs(y_test.values - y_pred)

# Display absolute errors for each prediction
for i in range(len(absolute_errors)):
    print(f"Actual: {y_test.iloc[i].values}, Predicted: {y_pred[i]}, Absolute Errors: {absolute_errors[i]}")
```
